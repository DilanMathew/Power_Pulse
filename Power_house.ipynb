```python
# Developing models:
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

features = [
    'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3',
    'Voltage', 'Global_reactive_power',
    'Rolling_Avg_3H', 'Rolling_Avg_24H', 'Daily_Avg_Global_active_power', 'Is_Peak_Hour', 'Hour'
]
target = 'Global_active_power'

X = df_n[features]
y = df_n[target]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#Evaluation
def evaluate_model(model, X_test, y_test, name):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} - MSE: {mse:.4f}, R²: {r2:.4f}")
    return mse, r2

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
evaluate_model(lr, X_test, y_test, "Linear Regression")

# Random Forest with Hyperparameter Tuning
rf_params = {'n_estimators': [50, 100], 'max_depth': [10, 20, None]}
rf = GridSearchCV(RandomForestRegressor(random_state=42), rf_params, cv=3, n_jobs=-1)
rf.fit(X_train, y_train)
evaluate_model(rf.best_estimator_, X_test, y_test, "Random Forest (Tuned)")

# Gradient Boosting with Hyperparameter Tuning
gb_params = {'n_estimators': [50, 100], 'learning_rate': [0.1, 0.05], 'max_depth': [3, 5]}
gb = GridSearchCV(GradientBoostingRegressor(random_state=42), gb_params, cv=3, n_jobs=-1)
gb.fit(X_train, y_train)
evaluate_model(gb.best_estimator_, X_test, y_test, "Gradient Boosting (Tuned)")

# Neural Network (MLPRegressor) with Hyperparameter Tuning
mlp_params = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'activation': ['relu', 'tanh'],
    'learning_rate_init': [0.001, 0.01]
}
mlp = GridSearchCV(MLPRegressor(random_state=42, max_iter=500), mlp_params, cv=3, n_jobs=-1)
mlp.fit(X_train, y_train)
evaluate_model(mlp.best_estimator_, X_test, y_test, "Neural Network (Tuned)")

#Compare Results
results = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest', 'Gradient Boosting', 'Neural Network'],
    'MSE': [
        mean_squared_error(y_test, lr.predict(X_test)),
        mean_squared_error(y_test, rf.best_estimator_.predict(X_test)),
        mean_squared_error(y_test, gb.best_estimator_.predict(X_test)),
        mean_squared_error(y_test, mlp.best_estimator_.predict(X_test))
    ],
    'R² Score': [
        r2_score(y_test, lr.predict(X_test)),
        r2_score(y_test, rf.best_estimator_.predict(X_test)),
        r2_score(y_test, gb.best_estimator_.predict(X_test)),
        r2_score(y_test, mlp.best_estimator_.predict(X_test))
    ]
})

print("\n Model Comparison:")
print(results)

# Comparison
plt.figure(figsize=(8, 5))
sns.barplot(data=results, x='R² Score', y='Model', palette='magma')
plt.title("Model R² Score Comparison")
plt.show()
```

Best Model: Random Forest (highest R², lowest MSE)
Neural Network and Gradient Boosting are very close in performance
Linear Regression is strong but slightly behind the ensemble metho

```python
# Predict on Test Data
y_pred_lr = lr.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_gb = gb.predict(X_test)
y_pred_mlp = mlp.predict(X_test)

# Plot
plt.figure(figsize=(16, 12))

# Linear Regression
plt.subplot(2, 2, 1)
sns.scatterplot(x=y_test, y=y_pred_lr, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Linear Regression: Actual vs Predicted')
plt.xlabel('Actual Global_active_power')
plt.ylabel('Predicted Global_active_power')

# Random Forest
plt.subplot(2, 2, 2)
sns.scatterplot(x=y_test, y=y_pred_rf, alpha=0.3, color='g')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Random Forest: Actual vs Predicted')
plt.xlabel('Actual Global_active_power')
plt.ylabel('Predicted Global_active_power')

# Gradient Boosting
plt.subplot(2, 2, 3)
sns.scatterplot(x=y_test, y=y_pred_gb, alpha=0.3, color='orange')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Gradient Boosting: Actual vs Predicted')
plt.xlabel('Actual Global_active_power')
plt.ylabel('Predicted Global_active_power')

# Neural Network
plt.subplot(2, 2, 4)
sns.scatterplot(x=y_test, y=y_pred_mlp, alpha=0.3, color='purple')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Neural Network: Actual vs Predicted')
plt.xlabel('Actual Global_active_power')
plt.ylabel('Predicted Global_active_power')

plt.tight_layout()
plt.show()
```
